```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = '/home/hello/jnotebook/prodigy/PRODIGY_DS_01/Students Performance.csv'
students_df = pd.read_csv(file_path)
```

```python
average_scores_by_year = students_df.groupby('Club_Join_Date').mean().reset_index()

plt.figure(figsize=(16, 10))

bar_width = 0.2
```

```python
r1 = range(len(average_scores_by_year))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]
r4 = [x + bar_width for x in r3]
```

```python
plt.bar(r1, average_scores_by_year['Math_Score'], color='b', width=bar_width, edgecolor='grey', label='Math Score')
plt.bar(r2, average_scores_by_year['Reading_Score'], color='g', width=bar_width, edgecolor='grey', label='Reading Score')
plt.bar(r3, average_scores_by_year['Writing_Score'], color='r', width=bar_width, edgecolor='grey', label='Writing Score')
plt.bar(r4, average_scores_by_year['Placement_Score'], color='y', width=bar_width, edgecolor='grey', label='Placement Score')



plt.xlabel('Club Join Date', fontweight='bold')
plt.ylabel('Average Score', fontweight='bold')
plt.title('Average Scores by Club Join Date', fontweight='bold')
plt.xticks([r + bar_width*1.5 for r in range(len(average_scores_by_year))], average_scores_by_year['Club_Join_Date'])




plt.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = '/home/hello/jnotebook/prodigy/PRODIGY_DS_01/Students Performance.csv'
students_df = pd.read_csv(file_path)
```

```python
plt.figure(figsize=(15, 10))
```

```python

# Math Score
plt.subplot(2, 2, 1)
plt.hist(students_df['Math_Score'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python

# Reading Score
plt.subplot(2, 2, 2)
plt.hist(students_df['Reading_Score'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Distribution of Reading Scores')
plt.xlabel('Reading Score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Writing Score
plt.subplot(2, 2, 3)
plt.hist(students_df['Writing_Score'], bins=10, color='salmon', edgecolor='black')
plt.title('Distribution of Writing Scores')
plt.xlabel('Writing Score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Placement Score
plt.subplot(2, 2, 4)
plt.hist(students_df['Placement_Score'], bins=10, color='yellow', edgecolor='black')
plt.title('Distribution of Placement Scores')
plt.xlabel('Placement Score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
